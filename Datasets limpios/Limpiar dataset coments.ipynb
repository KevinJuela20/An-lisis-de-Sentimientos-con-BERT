```python
import pandas as pd
import re
```

```python
def guardar_dataframe_como_csv(dataframe, nombre_archivo):
    # Utiliza el método to_csv para guardar el DataFrame en un archivo CSV
    dataframe.to_csv(nombre_archivo, index=False)  # El argumento index=False evita que se guarde el índice del DataFrame
```

```python
# Función para limpiar una frase
def limpiar_frase(frase):
    # Convertir a minúsculas
    frase = frase.lower()
    # Eliminar signos de puntuación y caracteres especiales
    frase_limpia = re.sub(r'[^a-zA-Z\s]', '', frase)
    return frase_limpia

def proceso_total(df):
    # Aplicar la función a la columna 'Frase' del DataFrame
    df['originalText'] = df['originalText'].apply(limpiar_frase)
    df['originalText'] = df['originalText'].str.strip()
    df = df[(df['sentiment'] != 'Neutral') & (df['originalText'] != '')]
    # Dataframe conformado por las columnas originalText y sentiment
    df_cleaned = df[['originalText', 'sentiment']]
    # Elinar datos vacíos
    df_cleaned = df_cleaned.dropna()
    # Eliminar filas completamente duplicadas
    df_cleaned = df_cleaned.drop_duplicates()
    return df_cleaned
```

# Dataset 1

```python
# Carga
df = pd.df = pd.read_csv('./jd-dataset-topics-Human-Review.csv')
print(df)
```

```python
# Limpieza total
df1_cleaned = proceso_total(df)
df1_cleaned
```
